```python
# trade matrix
trade_matrix_x = pd.DataFrame()
trade_matrix_m = pd.DataFrame()
countries = ['United States','China','Japan','Germany','India','United Kingdom','France','Italy','Canada', 'Korea, Rep.']
files = ['usa','china','japan','germany','india','uk','france','italy','canada','korea']
for i, file in enumerate(files):
    trade = pd.read_excel(f'{os.getcwd()}/trade/{file}.xlsx','Partner').iloc[:,:13]
    trade = trade[trade['Partner Name'].isin(countries)]
    trade = pd.concat([trade.iloc[:,:2], trade.iloc[:,-5:]], axis=1)
    
    x = []
    m = []
    for j, country in enumerate(countries):
        if (i == j):
            x.append(0)
            m.append(0)
        else:
            x.append(trade[trade['Partner Name'] == country]['Export (US$ Thousand)'].item())
            m.append(trade[trade['Partner Name'] == country]['Import (US$ Thousand)'].item())
    trade_matrix_x = trade_matrix_x.append([x])
    trade_matrix_m = trade_matrix_m.append([m])
            
trade_matrix_x = trade_matrix_x.to_numpy()
trade_matrix_m = trade_matrix_m.to_numpy().T
trade_matrix = pd.DataFrame((trade_matrix_x + trade_matrix_m)/2)
trade_matrix
```

```python
trade_matrix.to_csv(f'{os.getcwd()}/trade_matrix.csv', index=False)
```
